# result_network.ResultCatchment

``` python
result_network.ResultCatchment(self, catchment, res1d)
```

Class for wrapping a single ResultData catchment.

## Parameters

| Name      | Type            | Description                            | Default    |
|-----------|----------------|-----------------------------------|------------|
| catchment | IRes1DCatchment | MIKE 1D IRes1DCatchment object.        | *required* |
| res1d     | Res1D           | Res1D object the catchment belongs to. | *required* |

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [area](#mikeio1d.result_network.ResultCatchment.area) | The area of the catchment. |
| [derived_quantities](#mikeio1d.result_network.ResultCatchment.derived_quantities) | A list of available derived quantities. |
| [geometry](#mikeio1d.result_network.ResultCatchment.geometry) | A geometric representation of the catchment. Requires shapely. |
| [group](#mikeio1d.result_network.ResultCatchment.group) | The TimeSeriesIdGroup assosciated with this location. |
| [id](#mikeio1d.result_network.ResultCatchment.id) | The ID of the catchment. |
| [quantities](#mikeio1d.result_network.ResultCatchment.quantities) | A list of available quantities. |
| [res1d](#mikeio1d.result_network.ResultCatchment.res1d) | The Res1D instance that this location belongs to. |
| [res1d_catchment](#mikeio1d.result_network.ResultCatchment.res1d_catchment) | DHI.Mike1D.ResultDataAccess.IRes1DCatchment corresponding to this result location. |
| [type](#mikeio1d.result_network.ResultCatchment.type) | The type of the catchment. |

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [add_query](#mikeio1d.result_network.ResultCatchment.add_query) | Add a query to ResultNetwork.queries list. |
| [get_m1d_dataset](#mikeio1d.result_network.ResultCatchment.get_m1d_dataset) | Get IRes1DDataSet object associated with ResultCatchment. |
| [get_query](#mikeio1d.result_network.ResultCatchment.get_query) | Get a QueryDataCatchment for given data item. |
| [read](#mikeio1d.result_network.ResultCatchment.read) | Read the time series data for all quantities at this location into a DataFrame. |

### add_query

``` python
result_network.ResultCatchment.add_query(data_item)
```

### read

``` python

```

Read the time series data for all quantities at this location into a
DataFrame.

#### Parameters

| Name | Type | Description | Default |
|----|--------|---------------------------------------------------------|----|
| column_mode | str \| ColumnMode(optional) | Specifies the type of column index of returned DataFrame. ‘all’ - column MultiIndex with levels matching TimeSeriesId objects. ‘compact’ - same as ‘all’, but removes levels with default values. ‘timeseries’ - column index of TimeSeriesId objects | `None` |
