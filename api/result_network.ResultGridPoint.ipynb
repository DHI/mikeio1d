# result_network.ResultGridPoint

``` python
result_network.ResultGridPoint(
    self,
    reach,
    gridpoint,
    data_items,
    result_reach,
    res1d,
    tag='',
)
```

Class for wrapping a single ResultData grid point.

## Parameters

| Name | Type | Description | Default |
|--------|--------------|-------------------------------------------|-------|
| reach | IRes1DReach | MIKE 1D IRes1DReach object. | *required* |
| gridpoint | IRes1DGridPoint | MIKE 1D IRes1DGridPoint object. | *required* |
| data_items | list of IDataItem objects | A list of IDataItem objects (vector data object) the gridpoint has values defined on. | *required* |
| result_reach | ResultReach | Instance of ResultReach that this ResultGridPoint belongs to. | *required* |
| res1d | Res1D | Res1D object the grid point belongs to. | *required* |
| tag | str | Tag for reach location span where grid point belongs to. | `''` |

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [bottom_level](#mikeio1d.result_network.ResultGridPoint.bottom_level) | Bottom level of the gridpoint. |
| [chainage](#mikeio1d.result_network.ResultGridPoint.chainage) | Chainage of the gridpoint. |
| [derived_quantities](#mikeio1d.result_network.ResultGridPoint.derived_quantities) | A list of available derived quantities. |
| [gridpoint](#mikeio1d.result_network.ResultGridPoint.gridpoint) | IRes1DGridPoint corresponding to this result location. |
| [group](#mikeio1d.result_network.ResultGridPoint.group) | The TimeSeriesIdGroup assosciated with this location. |
| [quantities](#mikeio1d.result_network.ResultGridPoint.quantities) | A list of available quantities. |
| [reach](#mikeio1d.result_network.ResultGridPoint.reach) | Instance of ResultReach that this ResultGridPoint belongs to. |
| [reach_name](#mikeio1d.result_network.ResultGridPoint.reach_name) | Name of reach the gridpoint is on. |
| [res1d](#mikeio1d.result_network.ResultGridPoint.res1d) | The Res1D instance that this location belongs to. |
| [res1d_gridpoint](#mikeio1d.result_network.ResultGridPoint.res1d_gridpoint) | DHI.Mike1D.ResultDataAccess.IRes1DGridPoint corresponding to this result location. |
| [res1d_reach](#mikeio1d.result_network.ResultGridPoint.res1d_reach) | DHI.Mike1D.ResultDataAccess.IRes1DReach corresponding to this result location. |
| [result_reach](#mikeio1d.result_network.ResultGridPoint.result_reach) | Instance of ResultReach that this ResultGridPoint belongs to. |
| [xcoord](#mikeio1d.result_network.ResultGridPoint.xcoord) | X coordinate of the gridpoint. |
| [ycoord](#mikeio1d.result_network.ResultGridPoint.ycoord) | Y coordinate of the gridpoint. |

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [add_query](#mikeio1d.result_network.ResultGridPoint.add_query) | Add a query to ResultNetwork.queries list. |
| [get_m1d_dataset](#mikeio1d.result_network.ResultGridPoint.get_m1d_dataset) | Get IRes1DDataSet object associated with ResultGridPoint. |
| [get_query](#mikeio1d.result_network.ResultGridPoint.get_query) | Get a QueryDataReach for given data item. |
| [read](#mikeio1d.result_network.ResultGridPoint.read) | Read the time series data for all quantities at this location into a DataFrame. |

### add_query

``` python
result_network.ResultGridPoint.add_query(data_item)
```

### get_m1d_dataset

``` python
result_network.ResultGridPoint.get_m1d_dataset(m1d_dataitem=None)
```

Get IRes1DDataSet object associated with ResultGridPoint.

This is the reach IRes1DDataSet object because grid points do not have a
corresponding IRes1DDataSet object.

#### Parameters

| Name         | Type      | Description                  | Default |
|--------------|-----------|------------------------------|---------|
| m1d_dataitem | IDataItem | Ignored for ResultGridPoint. | `None`  |

#### Returns

| Name | Type          | Description                                           |
|--------|---------------|-------------------------------------------------|
|      | IRes1DDataSet | IRes1DDataSet object associated with ResultGridPoint. |

### get_query

``` python
result_network.ResultGridPoint.get_query(data_item)
```

### read

``` python
result_network.ResultGridPoint.read(column_mode=None)
```

Read the time series data for all quantities at this location into a
DataFrame.

#### Parameters

| Name | Type | Description | Default |
|----|--------|---------------------------------------------------------|----|
| column_mode | str \| ColumnMode(optional) | Specifies the type of column index of returned DataFrame. ‘all’ - column MultiIndex with levels matching TimeSeriesId objects. ‘compact’ - same as ‘all’, but removes levels with default values. ‘timeseries’ - column index of TimeSeriesId objects | `None` |
