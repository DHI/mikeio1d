# result_network.ResultReaches

``` python
result_network.ResultReaches(self, res1d)
```

Class for wrapping ResultData reaches.

By itself it is also a dict, which contains mapping between reach name
and IRes1DReach object or a list of IRes1DReach objects.

## Parameters

| Name  | Type  | Description                         | Default    |
|-------|-------|-------------------------------------|------------|
| res1d | Res1D | Res1D object the reaches belong to. | *required* |

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [derived_quantities](#mikeio1d.result_network.ResultReaches.derived_quantities) | A list of available derived quantities. |
| [group](#mikeio1d.result_network.ResultReaches.group) | The TimeSeriesIdGroup associated with these locations. |
| [locations](#mikeio1d.result_network.ResultReaches.locations) | A list of location objects (e.g. <ResultNode>). |
| [names](#mikeio1d.result_network.ResultReaches.names) | A list of location names (e.g. MUIDs). |
| [quantities](#mikeio1d.result_network.ResultReaches.quantities) | A list of available quantities. |
| [res1d](#mikeio1d.result_network.ResultReaches.res1d) | The Res1D instance that these locations belong to. |

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [read](#mikeio1d.result_network.ResultReaches.read) | Read the time series data for all quantities at these locations into a DataFrame. |
| [to_dataframe](#mikeio1d.result_network.ResultReaches.to_dataframe) | Read the time series data for all quantities at these locations into a DataFrame. |
| [to_geopandas](#mikeio1d.result_network.ResultReaches.to_geopandas) | Convert reaches to a geopandas.GeoDataFrame object. |

### read

``` python
result_network.ResultReaches.read(column_mode=None, include_derived=False)
```

Convert reaches to a geopandas.GeoDataFrame object.

By default, quantities are not included. To include quantities, use the
`agg` and `agg_kwargs` parameters.

#### Parameters

| Name | Type | Description | Default |
|-----|--------|------------------------------------------------------|------|
| agg | str or callable | Defines how to aggregate the quantities in time and space. Accepts any str or callable that is accepted by pandas.DataFrame.agg. | `None` |

#### Examples

    - 'mean'  : mean value of all quantities
    - 'max'   : maximum value of all quantities
    -  np.max : maximum value of all quantities

agg_kwargs : dict, default {} Aggregation function for specific column
levels (e.g. {time=‘min’, chainage=‘first’}). segmented : bool,
(default=True) True - one LineString per IRes1DReach object. False - one
LineString per reach name. include_derived : bool, default False Include
derived quantities.

#### Returns

| Name | Type | Description |
|-------|-------------------|----------------------------------------------|
| gdf | geopandas.GeoDataFrame | A GeoDataFrame object with reaches as LineString geometries. |

#### Examples

# Convert reaches to a GeoDataFrame (without quantities)

``` python
>>> gdf = res1d.reaches.to_geopandas()
```

# Convert reaches to a GeoDataFrame with aggregated quantities

``` python
>>> gdf = res1d.reaches.to_geopandas(agg='mean')
```
