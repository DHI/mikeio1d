# 

# result_network.ResultReach

``` python
result_network.ResultReach(self, reaches, res1d)
```

Class for wrapping a list of ResultData reaches having the same reach
name.

## Parameters

| Name | Type | Description | Default |
|-------|--------------|-------------------------------------------|---------|
| reaches | List\[IRes1DReach\] | A list of MIKE 1D IRes1DReach objects having the same reach name. | *required* |
| res1d | Res1D | Res1D object the reach belongs to. | *required* |

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [chainages](#mikeio1d.result_network.ResultReach.chainages) | List of chainages for the reach. |
| [derived_quantities](#mikeio1d.result_network.ResultReach.derived_quantities) | A list of available derived quantities. |
| [end_chainage](#mikeio1d.result_network.ResultReach.end_chainage) | End chainage of the reach. |
| [end_node](#mikeio1d.result_network.ResultReach.end_node) | End node of the reach. |
| [full_flow_discharge](#mikeio1d.result_network.ResultReach.full_flow_discharge) | Full flow discharge of the reach. |
| [geometry](#mikeio1d.result_network.ResultReach.geometry) | A geometric representation of the reach. Requires shapely. |
| [gridpoints](#mikeio1d.result_network.ResultReach.gridpoints) | List of gridpoints for the reach. |
| [group](#mikeio1d.result_network.ResultReach.group) | The TimeSeriesIdGroup assosciated with this location. |
| [height](#mikeio1d.result_network.ResultReach.height) | Height of the reach. |
| [length](#mikeio1d.result_network.ResultReach.length) | Length of the reach. |
| [n_gridpoints](#mikeio1d.result_network.ResultReach.n_gridpoints) | Number of gridpoints in the reach. |
| [name](#mikeio1d.result_network.ResultReach.name) | Name of the reach. |
| [quantities](#mikeio1d.result_network.ResultReach.quantities) | A list of available quantities. |
| [res1d](#mikeio1d.result_network.ResultReach.res1d) | The Res1D instance that this location belongs to. |
| [res1d_reaches](#mikeio1d.result_network.ResultReach.res1d_reaches) | List of DHI.Mike1D.ResultDataAccess.IRes1DReach corresponding to this result location. |
| [start_chainage](#mikeio1d.result_network.ResultReach.start_chainage) | Start chainage of the reach. |
| [start_node](#mikeio1d.result_network.ResultReach.start_node) | Start node of the reach. |

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [add_query](#mikeio1d.result_network.ResultReach.add_query) | Add a query to ResultNetwork.queries list. |
| [get_m1d_dataset](#mikeio1d.result_network.ResultReach.get_m1d_dataset) | Get IRes1DDataSet object associated with ResultReach. |
| [get_query](#mikeio1d.result_network.ResultReach.get_query) | Get a query for a data item. |
| [interpolate_reach_critical_level](#mikeio1d.result_network.ResultReach.interpolate_reach_critical_level) | Interpolate the critical level at a given chainage by linear interpolation from the bounding node critical levels. |
| [interpolate_reach_ground_level](#mikeio1d.result_network.ResultReach.interpolate_reach_ground_level) | Interpolate the ground level at a given chainage by linear interpolation from the bounding node ground levels. |
| [read](#mikeio1d.result_network.ResultReach.read) | Read the time series data for all quantities at this location into a DataFrame. |
| [to_dataframe](#mikeio1d.result_network.ResultReach.to_dataframe) | Read the time series data for all quantities at this location into a DataFrame. |

### add_query

``` python
result_network.ResultReach.add_query(data_item)
```

### get_m1d_dataset

``` python
result_network.ResultReach.get_m1d_dataset(m1d_dataitem=None)
```

Get IRes1DDataSet object associated with ResultReach.

A ResultReach may consist of several IRes1DDataSet objects. Therefore, a
IRes1DDataItem must be provided to identify the correct IRes1DDataSet.

#### Parameters

| Name | Type | Description | Default |
|-----------|---------|--------------------------------------------|---------|
| m1d_dataitem | IDataItem | The IRes1DDataItem associated with the returned IRes1DDataSet. | `None` |

#### Returns

| Name | Type          | Description                                       |
|------|---------------|---------------------------------------------------|
|      | IRes1DDataSet | IRes1DDataSet object associated with ResultReach. |

### get_query

``` python
result_network.ResultReach.get_query(data_item)
```

### read
