# Res1D

``` python
Res1D(
    self,
    file_path=None,
    reaches=None,
    nodes=None,
    catchments=None,
    time=None,
    step_every=None,
    quantities=None,
    derived_quantities=None,
    **kwargs,
)
```

Class for reading data from 1D network result files.

Currently supported formats are:

-   MIKE 1D network and catchment res1d files
-   MIKE 1D Long Term Statistics (LTS) res1d files
-   EPANET res, resx, and whr files generated by MIKE+
-   SWMM out files
-   MOUSE legacy prf, crf, and xrf files
-   MIKE 11 res11 files

## Parameters

| Name | Type | Description | Default |
|----------|----------------------|------------------------------------|------|
| file_path | str \| Path | File path of the result file. | `None` |
| reaches | list\[str\] \| None | Reach IDs to include when pre-loading dynamic results. None includes all. | `None` |
| nodes | list\[str\] \| None | Node IDs to include when pre-loading dynamic results. None includes all. | `None` |
| catchments | list\[str\] \| None | Catchment IDs to include when pre-loading dynamic results. None includes all. | `None` |
| time | Union\[tuple\[str\], list\[str\], slice, None\] | Start and end time of the data to read. Using None will read all data. | `None` |
| step_every | int \| None | Number specifying the time step frequency to output. None outputs all time steps. | `None` |
| quantities | list\[str\] \| None | Quantities to filter by (e.g. ‘WaterLevel’, ‘Discharge’). None includes all. | `None` |
| derived_quantities | list\[str\] \| None | Derived quantities to include when pre-loading dynamic results. None includes all. | `None` |

## Examples

Read all data from a res1d file into a Pandas DataFrame:

``` python
>>> res = Res1D("results.res1d")
>>> res.read()
```

Read all available data for nodes:

``` python
>>> res1d.nodes.read()
```

Read data for a specific node:

``` python
>>> res.nodes['node1'].read()
```

Read data for a specific quantity:

``` python
>>> res.nodes['node1'].WaterLevel.read()
```

Filter which data is pre-loaded, typically used for large files:

``` python
>>> nodes = ['node1', 'node2']
>>> reaches = ['reach1', 'reach2']
>>> times = slice('2020-01-01', '2020-01-02')
>>> res1d = Res1D('MyRes1D.res1d', nodes=nodes, reaches=reaches, time=times)
>>> res1d.read()
```

Only read every second time step:

``` python
>>> res1d = Res1D('MyRes1D.res1d', step_every=2)
>>> res1d.read()
```

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [catchments](#mikeio1d.Res1D.catchments) | Catchments of the result file. See [`ResultCatchments`](../api/result_network.ResultCatchments.html#mikeio1d.result_network.ResultCatchments). |
| [data](#mikeio1d.Res1D.data) | .NET object ResultData with the loaded res1d data. |
| [derived_quantities](#mikeio1d.Res1D.derived_quantities) | Derived quantities available for res1d file. |
| [end_time](#mikeio1d.Res1D.end_time) | End time of the result file. |
| [file_path](#mikeio1d.Res1D.file_path) | File path of the result file. |
| [global_data](#mikeio1d.Res1D.global_data) | Global data of the result file. See [`ResultGlobalDatas`](../api/result_network.ResultGlobalDatas.html#mikeio1d.result_network.ResultGlobalDatas). |
| [nodes](#mikeio1d.Res1D.nodes) | Nodes of the result file. See [`ResultNodes`](../api/result_network.ResultNodes.html#mikeio1d.result_network.ResultNodes). |
| [number_of_time_steps](#mikeio1d.Res1D.number_of_time_steps) | Number of time steps in the result file. |
| [projection_string](#mikeio1d.Res1D.projection_string) | Projection string of the result file. |
| [quantities](#mikeio1d.Res1D.quantities) | Quantities in res1d file. |
| [query](#mikeio1d.Res1D.query) | .NET object ResultDataQuery to use for querying the loaded res1d data. |
| [reaches](#mikeio1d.Res1D.reaches) | Reaches of the result file. See [`ResultReaches`](../api/result_network.ResultReaches.html#mikeio1d.result_network.ResultReaches). |
| [result_data](#mikeio1d.Res1D.result_data) | .NET object ResultData with the loaded res1d data. |
| [result_network](#mikeio1d.Res1D.result_network) | Deprecated. Use network property instead. |
| [result_type](#mikeio1d.Res1D.result_type) | Specifies what type of result file Res1D is. |
| [searcher](#mikeio1d.Res1D.searcher) | .NET object ResultDataSearcher to use for searching res1d data items on network. |
| [start_time](#mikeio1d.Res1D.start_time) | Start time of the result file. |
| [structures](#mikeio1d.Res1D.structures) | Structures of the result file. See `ResultStructures`. |
| [time_index](#mikeio1d.Res1D.time_index) | [`pandas.DatetimeIndex`](https://pandas.pydata.org/docs/reference/api/pandas.DatetimeIndex.html#pandas.DatetimeIndex) of the time index. |

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [add_derived_quantity](#mikeio1d.Res1D.add_derived_quantity) | Add a derived quantity to the Res1D object, propogating changes to the network. |
| [clear_queue](#mikeio1d.Res1D.clear_queue) | Clear the current active list of queries. |
| [extract](#mikeio1d.Res1D.extract) | Extract given queries to provided file. |
| [get_catchment_values](#mikeio1d.Res1D.get_catchment_values) | Get catchment values. Deprecated, use network.catchments instead. |
| [get_node_values](#mikeio1d.Res1D.get_node_values) | Get node values. Deprecated, use network.nodes instead. |
| [get_reach_end_values](#mikeio1d.Res1D.get_reach_end_values) | Get reach end values. Deprecated, use network.reaches instead. |
| [get_reach_start_values](#mikeio1d.Res1D.get_reach_start_values) | Get reach start values. Deprecated, use network.reaches instead. |
| [get_reach_sum_values](#mikeio1d.Res1D.get_reach_sum_values) | Get reach sum values. Deprecated, use network.reaches instead. |
| [get_reach_value](#mikeio1d.Res1D.get_reach_value) | Get reach value. Deprecated, use network.reaches instead. |
| [get_reach_values](#mikeio1d.Res1D.get_reach_values) | Get reach values. Deprecated, use network.reaches instead. |
| [get_supported_file_extensions](#mikeio1d.Res1D.get_supported_file_extensions) | Get supported file extensions for Res1D. |
| [info](#mikeio1d.Res1D.info) | Print information about the result file. |
| [merge](#mikeio1d.Res1D.merge) | Merge res1d files. |
| [modify](#mikeio1d.Res1D.modify) | Modify the result data. |
| [read](#mikeio1d.Res1D.read) | Read result data into a pandas DataFrame. |
| [read_all](#mikeio1d.Res1D.read_all) | Read all data from res1d file to dataframe. Deprecated, use read() instead. |
| [remove_derived_quantity](#mikeio1d.Res1D.remove_derived_quantity) | Remove a derived quantity from the Res1D object, propogating changes to the network. |
| [save](#mikeio1d.Res1D.save) | Save the ResultData to a new res1d file. |
| [to_csv](#mikeio1d.Res1D.to_csv) | Extract to csv file. |
| [to_dfs0](#mikeio1d.Res1D.to_dfs0) | Extract to dfs0 file. |
| [to_txt](#mikeio1d.Res1D.to_txt) | Extract to txt file. |

### add_derived_quantity

``` python
Res1D.add_derived_quantity(derived_quantity)
```

Extract given queries to provided file.

File type is determined from file_path extension. The supported formats
are:

-   csv
-   dfs0
-   txt

#### Parameters

| Name | Type | Description | Default |
|----------|---------------------------|-------------------------------|-----|
| file_path | str | Output file path. | *required* |
| queries | Optional\[List\[QueryData\] \| QueryData \| List\[TimeSeriesId\] \| TimeSeriesId\] | A list of queries to read. If None, all data is read. | `None` |
| time_step_skipping_number | int | Number specifying the time step frequency to output. | `1` |
| ext | str | Output file type to use instead of determining it from extension. Can be ‘csv’, ‘dfs0’, ‘txt’. | `None` |

#### Notes

The `queries` parameter exists for historical reasons where queries were
built manually. Now, this is handled automatically
e.g. `res.nodes['node1'].WaterLevel.read()`. It may still be useful so
it has been kept for now.

Merge res1d files.

It is possible to merge three kinds of result files:

-   Regular res1d (HD, RR, etc.)
-   LTS extreme statistics
-   LTS chronological statistics

For regular res1d files the requirement is that the simulation start
time of the first file matches the simulation end time of the second
file (the same principle for subsequent files).

For LTS result files, meaningful merged result file is obtained when
simulation periods for the files do not overlap.

#### Parameters

| Name | Type | Description | Default |
|------------|-------------------|----------------------------------|--------|
| file_names | list of str or Res1D objects | List of res1d file names to merge. | *required* |
| merged_file_name | str | File name of the res1d file to store the merged data. | *required* |

### modify

``` python
Res1D.modify(data_frame, file_path=None)
```

Modify the result data.

Note that this only modifies the data in memory, not the original file.
To save the modified data to a new file, use
[`save`](../api/Res1D.html#mikeio1d.Res1D.save). Alternatively, use
`file_path` to save the modified data to a new file.

#### Parameters

| Name | Type | Description | Default |
|-------|---------|--------------------------------------------------|-------|
| data_frame | pandas.DataFrame | [`pandas.DataFrame`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html#pandas.DataFrame) with the new data, with columns matching data generated by [`read`](../api/Res1D.html#mikeio1d.Res1D.read). | *required* |
| file_path | str | File path for the new res1d file. Optional. | `None` |

#### Notes

This modifies the .NET ResutlData’s TimeData object.

#### Examples

``` python
>>> res = Res1D('results.res1d')
>>> data = res.read()
>>> data += 100
>>> res.modify(data)
```
