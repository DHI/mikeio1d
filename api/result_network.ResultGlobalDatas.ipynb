# result_network.ResultGlobalDatas

``` python
result_network.ResultGlobalDatas(self, res1d)
```

Class for wrapping ResultData global data items.

By itself it is also a dict, which contains mapping between global data
item quantity ID and IDataItem object.

## Parameters

| Name  | Type  | Description                            | Default    |
|-------|-------|----------------------------------------|------------|
| res1d | Res1D | Res1D object the catchments belong to. | *required* |

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [derived_quantities](#mikeio1d.result_network.ResultGlobalDatas.derived_quantities) | A list of available derived quantities. |
| [group](#mikeio1d.result_network.ResultGlobalDatas.group) | The TimeSeriesIdGroup associated with these locations. |
| [locations](#mikeio1d.result_network.ResultGlobalDatas.locations) | A list of location objects (e.g. <ResultNode>). |
| [names](#mikeio1d.result_network.ResultGlobalDatas.names) | A list of location names (e.g. MUIDs). |
| [quantities](#mikeio1d.result_network.ResultGlobalDatas.quantities) | A list of available quantities. |
| [res1d](#mikeio1d.result_network.ResultGlobalDatas.res1d) | The Res1D instance that these locations belong to. |

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [read](#mikeio1d.result_network.ResultGlobalDatas.read) | Read the time series data for all quantities at these locations into a DataFrame. |

### read

``` python
result_network.ResultGlobalDatas.read(column_mode=None, include_derived=False)
```

Read the time series data for all quantities at these locations into a
DataFrame.

#### Parameters

| Name | Type | Description | Default |
|-----|--------|--------------------------------------------------------|----|
| column_mode | str \| ColumnMode(optional) | Specifies the type of column index of returned DataFrame. ‘all’ - column MultiIndex with levels matching TimeSeriesId objects. ‘compact’ - same as ‘all’, but removes levels with default values. ‘timeseries’ - column index of TimeSeriesId objects | `None` |
| include_derived | bool | Include derived quantities. | `False` |
