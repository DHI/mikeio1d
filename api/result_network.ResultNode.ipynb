# result_network.ResultNode

``` python
result_network.ResultNode(self, node, res1d)
```

Class for wrapping a single ResultData node.

## Parameters

| Name  | Type       | Description                       | Default    |
|-------|------------|-----------------------------------|------------|
| node  | IRes1DNode | MIKE 1D IRes1DNode object.        | *required* |
| res1d | Res1D      | Res1D object the node belongs to. | *required* |

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [bottom_level](#mikeio1d.result_network.ResultNode.bottom_level) | Bottom level of the node. |
| [critical_level](#mikeio1d.result_network.ResultNode.critical_level) | Critical level of the node. |
| [derived_quantities](#mikeio1d.result_network.ResultNode.derived_quantities) | A list of available derived quantities. |
| [diameter](#mikeio1d.result_network.ResultNode.diameter) | Diameter of the node. |
| [geometry](#mikeio1d.result_network.ResultNode.geometry) | A geometric representation of the node. Requires shapely. |
| [ground_level](#mikeio1d.result_network.ResultNode.ground_level) | Ground level of the node. |
| [group](#mikeio1d.result_network.ResultNode.group) | The TimeSeriesIdGroup assosciated with this location. |
| [id](#mikeio1d.result_network.ResultNode.id) | Node ID. |
| [quantities](#mikeio1d.result_network.ResultNode.quantities) | A list of available quantities. |
| [res1d](#mikeio1d.result_network.ResultNode.res1d) | The Res1D instance that this location belongs to. |
| [res1d_node](#mikeio1d.result_network.ResultNode.res1d_node) | DHI.Mike1D.ResultDataAccess.IRes1DNode corresponding to this result location. |
| [type](#mikeio1d.result_network.ResultNode.type) | Node type. |
| [xcoord](#mikeio1d.result_network.ResultNode.xcoord) | X coordinate of the node. |
| [ycoord](#mikeio1d.result_network.ResultNode.ycoord) | Y coordinate of the node. |

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [add_query](#mikeio1d.result_network.ResultNode.add_query) | Add a query to ResultNetwork.queries list. |
| [get_m1d_dataset](#mikeio1d.result_network.ResultNode.get_m1d_dataset) | Get IRes1DDataSet object associated with ResultNode. |
| [get_query](#mikeio1d.result_network.ResultNode.get_query) | Get a QueryDataNode for given data item. |
| [read](#mikeio1d.result_network.ResultNode.read) | Read the time series data for all quantities at this location into a DataFrame. |

### add_query

``` python
result_network.ResultNode.add_query(data_item)
```

### read

``` python

```

Read the time series data for all quantities at this location into a
DataFrame.

#### Parameters

| Name | Type | Description | Default |
|----|--------|---------------------------------------------------------|----|
| column_mode | str \| ColumnMode(optional) | Specifies the type of column index of returned DataFrame. ‘all’ - column MultiIndex with levels matching TimeSeriesId objects. ‘compact’ - same as ‘all’, but removes levels with default values. ‘timeseries’ - column index of TimeSeriesId objects | `None` |
