### Read network results

Use [`mikeio1d.open`](./api/open.html#mikeio1d.open) to create a
[Res1D](user-guide/res1d.qmd) object, then
[`read`](./api/Res1D.html#mikeio1d.Res1D.read) its content into a
[`DataFrame`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html#pandas.DataFrame).

````python
import mikeio1d

res = mikeio1d.open("data/network.res1d")
df = res.read()
df.head()
````

Request more specific data by using the [Res1D](user-guide/res1d.qmd)
object to select specific [Locations](user-guide/locations.qmd).

````python
df = res.reaches['100l1'].read()
df.head()
````

Similarly, select a specific [Quantity](user-guide/quantities.qmd) that
you can both
[`read`](./api/result_network.ResultQuantity.html#mikeio1d.result_network.ResultQuantity.read)
and
[`plot`](./api/result_network.ResultQuantity.html#mikeio1d.result_network.ResultQuantity.plot).

````python
res.nodes['100'].WaterLevel.plot()
````

### Read cross section data

Use [`mikeio1d.open`](./api/open.html#mikeio1d.open) to create an
[Xns11](user-guide/xns11.qmd) object, then get a
[`DataFrame`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html#pandas.DataFrame)
overview of its content.

````python
xns = mikeio1d.open("data/mikep_cs_demo.xns11")
xns.to_dataframe()
````

Select a specific
[`CrossSection`](./api/cross_sections.CrossSection.html#mikeio1d.cross_sections.CrossSection)
using [`sel`](./api/Xns11.html#mikeio1d.Xns11.sel), then plot it with
[`plot`](./api/cross_sections.CrossSection.html#mikeio1d.cross_sections.CrossSection.plot).

````python
xs = xns.sel(location_id="basin_left1", chainage=122.042, topo_id='1')
xs.plot()
````
