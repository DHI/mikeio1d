# Overview

##  Res1D

[Res1D](res1d.qmd) is the main interface for accessing data on a
[Network](network.qmd).

```python
import mikeio1d
res = mikeio1d.open('../data/network.res1d')
res
```

##  Locations

[Locations](locations.qmd) are where model results exist in the network.
The main location types are
[nodes](../api/result_network.ResultNodes.html#mikeio1d.result_network.ResultNodes),
[reaches](../api/result_network.ResultReaches.html#mikeio1d.result_network.ResultReaches),
and
[catchments](../api/result_network.ResultCatchments.html#mikeio1d.result_network.ResultCatchments).
Each of these location types are accessible via a
[`Res1D`](../api/Res1D.html#mikeio1d.Res1D) object.

```python
res.nodes
```

Specific
[locations](../api/result_network.ResultLocation.html#mikeio1d.result_network.ResultLocation)
are accessible via their unique ID.

```python
res.nodes['101']
```

##  Quantities

[Quantities](quantities.qmd) are the actual model results, which can be
read into a
[`DataFrame`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html#pandas.DataFrame)
or plotted directly.

```python
df = res.nodes['101'].WaterLevel.read()
df.head()
```

```python
res.nodes['101'].WaterLevel.plot()
```

[Quantity
collections](../api/result_network.ResultQuantityCollection.html#mikeio1d.result_network.ResultQuantityCollection)
enable reading a specific quantity for many locations at once.

```python
df = res.reaches.Discharge.read()
df.describe().T.head()
```

##  Xns11

[Xns11](xns11.qmd) is the main interface for accessing cross section
data.

```python
xns = mikeio1d.open('../data/mikep_cs_demo.xns11')
xns['basin_left1','122.042','1'].plot(with_marker_labels=False)
```

Access a cross section’s
[`raw`](../api/cross_sections.CrossSection.html#mikeio1d.cross_sections.CrossSection.raw)
info as a
[`DataFrame`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html#pandas.DataFrame).

```python
df = xns['basin_left1','122.042','1'].raw
df.head()
```

Access a cross section’s
[`processed`](../api/cross_sections.CrossSection.html#mikeio1d.cross_sections.CrossSection.processed)
info as a
[`DataFrame`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html#pandas.DataFrame).

```python
df = xns['basin_left1','122.042','1'].processed
df.head()
```
